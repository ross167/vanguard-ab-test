```python
import pandas as pd
```

```python
df_pt1 = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/raw/df_final_web_data_pt_1.csv')
df_pt1
```

```python
df_pt2 = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/raw/df_final_web_data_pt_2.csv')
df_pt2
```

```python
df_pt = pd.concat([df_pt1,df_pt2])
df_pt
```

```python
df_pt.duplicated().sum()
df_pt = df_pt.drop_duplicates().reset_index().drop(columns="index")
df_pt
```

```python
df_exp = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/clean/df_final_experiment_clients_cleaned.csv')
```

```python
df_exp.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df_exp.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
df_exp
```

```python
df_pt["date_time"] = pd.to_datetime(df_pt["date_time"])
```

```python
df_pt = df_pt.sort_values(by=["client_id","visitor_id","visit_id","date_time"])
```

```python
df = df_pt.merge(df_exp, on="client_id", how="left")
```

```python
df.dropna(inplace=True)
```

```python
df['variation'].unique()
```

```python
df_test = df[df['variation'] == "Test"]
df_test
```

```python
df_control = df[df['variation'] == "Control"]
df_control
```

```python
df_test[df_test['process_step'] == 'confirm']['visitor_id'].nunique()
```

```python
#Calculate the number of visitors that reach confirm step:

number_of_clients_confirm_testgroup = df_test[df_test['process_step'] == 'confirm']['visitor_id'].nunique()
number_of_clients_confirm_controlgroup = df_control[df_control['process_step'] == 'confirm']['visitor_id'].nunique()

#Calculate the total number of visitors in each group:
total_number_of_clients_testgroup = df_test['client_id'].nunique()
total_number_of_clients_controlgroup = df_control['client_id'].nunique()

#Calculate the completion rate for each group:
conversion_rate_testgroup = number_of_clients_confirm_testgroup / total_number_of_clients_testgroup
conversion_rate_controlgroup = number_of_clients_confirm_controlgroup / total_number_of_clients_controlgroup
print(f"Completion rate for Test group: {conversion_rate_testgroup:.2%}")
print(f"Completion rate for Control group: {conversion_rate_controlgroup:.2%}")
print(f"Test group has a {(conversion_rate_testgroup - conversion_rate_controlgroup):.2%} higher completion rate than Control group")
```
