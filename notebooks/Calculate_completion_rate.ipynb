```python
import pandas as pd
```

```python
df_pt1 = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/raw/df_final_web_data_pt_1.csv')
df_pt1
```

```python
df_pt2 = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/raw/df_final_web_data_pt_2.csv')
df_pt2
```

```python
df_pt = pd.concat([df_pt1,df_pt2])
df_pt
```

```python
df_pt.duplicated().sum()
df_pt = df_pt.drop_duplicates().reset_index().drop(columns="index")
df_pt
```

```python
df_exp = pd.read_csv('https://raw.githubusercontent.com/ross167/vanguard-ab-test/refs/heads/main/database/clean/df_final_experiment_clients_cleaned.csv')
```

```python
df_exp.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df_exp.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
df_exp
```

```python
df_pt["date_time"] = pd.to_datetime(df_pt["date_time"])
```

```python
df = df_pt.merge(df_exp, on="client_id", how="left")
```

```python
df.dropna(inplace=True)
```

```python
step = {
    "start": 0,
    "step_1": 1,
    "step_2": 2,
    "step_3": 3,
    "confirm": 4}
```

```python
df['process_step'] = df['process_step'].map(step)
```

```python
df
```

```python
df_test = df[df['variation'] == "Test"]
df_test
```

```python
df_control = df[df['variation'] == "Control"]
df_control
```

```python
df_test[df_test['process_step'] == 4]['visitor_id'].nunique()
```

```python
#Function that calculate the completion rate by different parameters and step

def calculate_completion_rate(param,step):
    number_of_clients_confirm_testgroup = df_test[df_test['process_step'] == step][param].nunique()
    number_of_clients_confirm_controlgroup = df_control[df_control['process_step'] == step][param].nunique()

    #Calculate the total number of visitors in each group:

    total_number_of_clients_testgroup = df_test[df_test['process_step'] == 0][param].nunique()
    total_number_of_clients_controlgroup = df_control[df_control['process_step'] == 0][param].nunique()

    #Calculate the completion rate for each group:

    completion_rate_testgroup = round((number_of_clients_confirm_testgroup / total_number_of_clients_testgroup)*100, 2)
    completion_rate_controlgroup = round((number_of_clients_confirm_controlgroup / total_number_of_clients_controlgroup)*100, 2)


    #print(f"Successful rate of step number {step} for Test group: {completion_rate_testgroup}%")
    #print(f"Successful rate of step number {step} for Control group: {completion_rate_controlgroup}%")
    #print(f"Test group has a {(conversion_rate_testgroup - conversion_rate_controlgroup):.2%} higher completion rate for step {step} than Control group counting by {parram}")

    return  completion_rate_testgroup, completion_rate_controlgroup
```

```python
calculate_completion_rate('client_id',4)
```

```python
#Success rate by each group count by client_id 
df_client_id = pd.DataFrame([calculate_completion_rate('client_id', step) for step in range (5)], columns=["test_group", "control_group"]).reset_index()
df_client_id  = df_client_id.rename(columns = {'index': 'step'})
df_client_id 
```

```python
#Success rate by each group count by visitor_id (Each client has average 2 visitor_id)
df_visitor_id = pd.DataFrame([calculate_completion_rate('visitor_id', step) for step in range (5)], columns=["test_group", "control_group"]).reset_index()
df_visitor_id  = df_visitor_id.rename(columns = {'index': 'step'})
df_visitor_id 
```

```python
df_visit_id = pd.DataFrame([calculate_completion_rate('visit_id', step) for step in range (5)], columns=["test_group", "control_group"]).reset_index()
df_visit_id = df_visit_id.rename(columns = {'index': 'step'})
df_visit_id 
```
