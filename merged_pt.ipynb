```python
df_pt.duplicated().sum()
df_pt = df_pt.drop_duplicates().reset_index().drop(columns="index")
df_pt
```

```python
df_pt["date_time"] = pd.to_datetime(df_pt["date_time"])
```

```python
df_pt = df_pt.sort_values(by=["visit_id","date_time"])
```

```python
df_pt["duration"] = ""
```

```python
def duration_calculator(row, df):
    if row["process_step"] == "start":
        return 0
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 0
    prev_row = df.iloc[prev_idx]
    return int((row["date_time"] - prev_row["date_time"]).total_seconds())
```

```python

df_pt["duration"] = df_pt.apply(duration_calculator, axis=1, df=df_pt)
```

```python
#execution of each step:
order = ["start","step_1","step_2","step_3","confirm"]
exe = df_pt.groupby("process_step").agg({"visit_id":"nunique"}).sort_values(by="process_step").reindex(order)

```

```python

def s_rate(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[prev_idx]
    return round((row["visit_id"] / prev_row["visit_id"])*100,2)
```

```python
exe["step_success(%)"] = exe.apply(s_rate, axis=1, df=exe) 
exe
```

```python
def a_s_rate(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[0]
    return round((row["visit_id"] / prev_row["visit_id"])*100,2)
```

```python
exe["overall_success(%)"] = exe.apply(a_s_rate, axis=1, df=exe) 
exe
```

```python
df_exp = pd.read_csv(r"C:\Users\rcrds\Desktop\Ironhack\Projects\Vanguard AB Test\vanguard-ab-test\database\clean\df_final_experiment_clients_cleaned.csv")
df_exp
```

```python
df_pt_new = pd.merge(df_pt,df_exp,how="left",on="client_id").drop(columns="Unnamed: 0")
df_pt_new.isnull().sum()
```

```python
df_pt_new = df_pt_new.dropna(thresh=7)
```

```python
#error detection

df_error = df_pt_new.pivot_table(index="visit_id", columns="process_step", values = "client_id", aggfunc="count")
df_error = df_error.fillna(0)
df_error
```

```python
def error_checking(row):
    if row["start"] == 0:
        return "error"
    elif row["step_1"] == 0 and (row["step_2"]+row["step_3"]+row["confirm"] > 0):
        return "error"
    elif row["step_2"] == 0 and (row["step_3"]+row["confirm"] > 0):
        return "error"
    elif row["step_3"] == 0 and row["confirm"] > 0:
        return "error"
    else:
        return "valid"
```

```python
df_error["error_valid"] = df_error.apply(error_checking,axis=1)
df_error
```

```python
df_pt_new = pd.merge(df_pt_new,df_error,how="left",on="visit_id")
df_pt_new
```

```python

df_pt_new = df_pt_new[df_pt_new["error_valid"] == "valid"]
df_pt_new
```

```python
df_pt_new["repetition"] = df_pt_new["visit_id"].duplicated()
df_pt_new
```

```python
def error2(row):
    if row["process_step"] != "start" and row["repetition"] == False:
        return "error"
    else:
        return "valid"
```

```python
df_pt_new["erro2"] = df_pt_new.apply(error2, axis=1)
df_pt_new
```

```python
exe2 = df_pt_new.pivot_table(index="process_step", columns="Variation", values="client_id", aggfunc="nunique").reindex(order)
exe2
```

```python
def s_rate2(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[prev_idx]
    return round((row["Control"] / prev_row["Control"])*100,2)

def a_s_rate2(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[0]
    return round((row["Control"] / prev_row["Control"])*100,2)
```

```python
exe2["control_step_completion(%)"] = exe2.apply(s_rate2, axis=1, df=exe2)
exe2
```

```python
exe2["control_overall_completion(%)"] = exe2.apply(a_s_rate2, axis=1, df=exe2)
exe2
```

```python
def s_rate3(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[prev_idx]
    return round((row["Test"] / prev_row["Test"])*100,2)

def a_s_rate3(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[0]
    return round((row["Test"] / prev_row["Test"])*100,2)
```

```python
exe2["test_step_completion(%)"] = exe2.apply(s_rate3, axis=1, df=exe2)
exe2
```

```python
exe2["test_overall_completion(%)"] = exe2.apply(a_s_rate3, axis=1, df=exe2)
exe2
```

```python
exe2 = exe2.reindex(columns=["Control", "control_step_completion(%)", "control_overall_completion(%)", "Test", "test_step_completion(%)", "test_overall_completion(%)"])
exe2
```

```python
def dif_step(row):
    result = row["test_step_completion(%)"] - row["control_step_completion(%)"]
    return result

def dif_overall(row):
    result = row["test_overall_completion(%)"] - row["control_overall_completion(%)"]
    return result
```

```python
exe2["diference_steps"] = exe2.apply(dif_step,axis=1)
exe2
```

```python
exe2["diference_overall"] = exe2.apply(dif_overall,axis=1)
exe2
```
