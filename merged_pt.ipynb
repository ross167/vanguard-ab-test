```python
df_pt1 = pd.read_csv(r"C:\Users\rcrds\Downloads\df_final_web_data_pt_1.txt")
df_pt1
```

```python
df_pt2 = pd.read_csv(r"C:\Users\rcrds\Downloads\df_final_web_data_pt_2.txt")
df_pt2
```

```python
df_pt = pd.concat([df_pt1,df_pt2])
df_pt
```

```python
df_pt.duplicated().sum()
df_pt = df_pt.drop_duplicates().reset_index().drop(columns="index")
df_pt
```

```python
df_pt["date_time"] = pd.to_datetime(df_pt["date_time"])
```

```python
df_pt.process_step.unique()
```

```python
df_pt.isnull().sum()
```

```python
df_pt = df_pt.sort_values(by=["client_id","visitor_id","visit_id","date_time"])
```

```python
df_pt["duration"] = ""
```

```python
def duration_calculator(row, df):
    if row["process_step"] == "start":
        return 0
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 0
    prev_row = df.iloc[prev_idx]
    return int((row["date_time"] - prev_row["date_time"]).total_seconds())
```

```python

df_pt["duration"] = df_pt.apply(duration_calculator, axis=1, df=df_pt)
```

```python
#execution of each step:
order = ["start","step_1","step_2","step_3","confirm"]
exe = df_pt.groupby("process_step").agg({"visit_id":"nunique"}).sort_values(by="process_step").reindex(order)

```

```python

def s_rate(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[prev_idx]
    return round((row["visit_id"] / prev_row["visit_id"])*100,2)
```

```python
exe["step_success(%)"] = exe.apply(s_rate, axis=1, df=exe) 
exe
```

```python
def a_s_rate(row,df):
    prev_idx = df.index.get_loc(row.name) - 1
    if prev_idx < 0:
        return 100
    prev_row = df.iloc[0]
    return round((row["visit_id"] / prev_row["visit_id"])*100,2)
```

```python
exe["overall_success(%)"] = exe.apply(a_s_rate, axis=1, df=exe) 
exe
```
